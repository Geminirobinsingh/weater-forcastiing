```python
def updateEd(Weather):
    if Weather in ['Fog','Haze','Freezing Drizzle,Fog','Freezing Rain,Fog','Freezing Fog','Rain,Fog', 'Drizzle,Fog','Snow,Fog','Thunderstorms,Rain Showers,Fog','Rain,Snow,Fog','Rain,Snow,Fog','Freezing Rain,Ice Pellets,Fog','Drizzle,Snow,Fog','Drizzle,Snow,Fog''Rain Showers,Fog']:
        return 'fog'
    elif Weather in ['Mostly Cloudy','Cloudy','Freezing Rain','Freezing Drizzle','Moderate Snow',]:
        return 'cloudy'
    elif Weather in['Rain', 'Drizzle' 'Rain Showers','Freezing Drizzle,Haze','Freezing Rain,Haze','Rain,Haze','Thunderstorms,Rain', 'Thunderstorms,Rain Showers', 'Thunderstorms,Heavy Rain Showers','Thunderstorms','Thunderstorms,Moderate Rain Showers,Fog','Moderate Rain,Fog','Rain,Ice Pellets','Drizzle,Snow','Rain,Snow Grains','Rain,Snow,Ice Pellets',]:
        return 'rain'
    elif  Weather in['Mainly Clear','Clear', ]:
        return 'clear'
    elif  Weather in['Snow Showers', 'Snow', 'Rain,Snow', 'Moderate Snow', 'Freezing Drizzle,Snow','Snow,Haze','Drizzle,Snow','Snow Showers,Fog', 'Moderate Snow,Blowing Snow','Snow Pellets',]:
        return 'snow'
    else:
        return 'other'
   
```

```python
rs['Weather'] =rs['Weather'].apply(updateEd)
rs
```

```python
X = rs.drop(['Date/Time', 'Weather'], axis=1)
y = rs['Weather']
```

```python
X
```

```python
from sklearn.model_selection import cross_val_score
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
rs.shape
```

```python
ks = list(range(1, 400, 2))
ks
```

```python
cvScores = []
import warnings

warnings.filterwarnings("ignore")
```

```python
for k in ks:
    knn = KNeighborsClassifier(n_neighbors=k)
    score = cross_val_score(knn, X, y, cv=5, scoring='accuracy').mean()
    print(f"Accuracy: {score} with {k} neighbors")
    cvScores.append(score)
```

```python
max(cvScores)
```

```python
cvScores.index(max(cvScores))
```

```python
finalKNN = KNeighborsClassifier(n_neighbors=117)
finalKNN.fit(X, y)
```
